```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
los_angeles = df[df['City']=="Los Angeles"]
selected_columns = ['Address', 'Sold Price', 'Type', 'Year built', 'Heating', 'Cooling', 'Parking', 'Bedrooms', 'Bathrooms',
                    'Elementary School', 'Elementary School Score', 'Elementary School Distance', 'Middle School', 'Middle School Score', 'Middle School Distance',
                    'High School', 'High School Score', 'High School Distance', 'Appliances included', 'Laundry features', 'Tax assessed value',
                    'Annual tax amount', 'City']
los_angeles = los_angeles[selected_columns]
los_angeles = los_angeles[(los_angeles['Type'] != "VacantLand") & (los_angeles['Type'] != "Unknown") & (los_angeles['Type'] != "Cooperative")]
```

```python
import pandas as pd
import tiktoken

# Ensure all columns are combined into a single text (adjust columns as needed)
los_angeles['joined_text'] = los_angeles.fillna('').astype(str).agg(' '.join, axis=1)

# Choose the tokenizer for your target OpenAI model
# Replace 'gpt-3.5-turbo' with 'gpt-4' if needed
tokenizer = tiktoken.get_encoding("cl100k_base")

# Count tokens for each row
los_angeles['token_count'] = los_angeles['joined_text'].apply(lambda x: len(tokenizer.encode(x)))

# Print total tokens
total_tokens = los_angeles['token_count'].sum()
print("Total Tokens:", total_tokens)

# If you want to see the token counts per row
print(los_angeles[['joined_text', 'token_count']])
```
